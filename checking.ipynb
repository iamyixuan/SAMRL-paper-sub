```python
import glob
import pandas as pd
import numpy as np
```

```python
folder = './eval_plots/paper_plots/agent_viz/test/extraStates/'
files = glob.glob(folder + '*.csv')
dfs = []
for f in files:
    df_tmp = pd.read_csv(f)
    dfs.append(df_tmp)
dfs = pd.concat(dfs)
dfs.describe()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10, 10))
ax = plt.gca()
dfs.hist(bins=50, ax=ax)
plt.tight_layout()
```
